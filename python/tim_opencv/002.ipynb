```python
import itertools
import random


img = cv2.imread('assets/logo.jpg')
img2 = img.copy()

for i, j in itertools.product(range(20), range(img.shape[1])):
    img2[i, j] = [random.randint(0, 255), random.randint(0, 255), random.randint(0, 255)]

show_image(img2, 'Image')
```
